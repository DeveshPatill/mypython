```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram 
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Step 1: Define the dataset
data = np.array([[1], [5], [8], [10], [19], [20]])

# Step 2: Compute the linkage matrix for hierarchical clustering
linkage_matrix = linkage(data, method='ward')

# Step 3: Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix, labels=['1', '5', '8', '10', '19', '20'])
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Step 4: Perform agglomerative clustering
model = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels = model.fit_predict(data)

# Step 5: Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data, np.zeros_like(data), c=labels, cmap='viridis', s=100)
for i, txt in enumerate(['1', '5', '8', '10', '19', '20']):
    plt.annotate(txt, (data[i], 0))
plt.title('Hierarchical Clustering Results')
plt.xlabel('Data Points')
plt.ylabel('Y (Fixed for Visualization)')
plt.show()

# Step 6: Print the clusters
print("Cluster Assignments:", labels)
```
